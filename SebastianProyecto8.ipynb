```python
import matplotlib.pyplot as plt

plt.close('all')

x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y1 = [1, 3, 5, 3, 1, 3, 5, 3, 1]
y2 = [2, 4, 6, 4, 2, 4, 6, 4, 2]

plt.figure(figsize=(7, 4))
plt.plot(x, y1, label="line I")
plt.plot(x, y2, label="line H")

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line Graph Example")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.close('all')

x1 = [1, 3, 4, 5, 6, 7, 9]
y1 = [4, 7, 2, 4, 7, 8, 3]

x2 = [2, 4, 6, 8, 10]
y2 = [5, 6, 2, 6, 2]

plt.figure(figsize=(7, 4))
plt.bar(x1, y1, label="Azul", color="blue")
plt.bar(x2, y2, label="Verde", color="green")

plt.xlabel("bar number")
plt.ylabel("bar height")
plt.title("Bar Chart Example")
plt.legend()
plt.grid(axis="y")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.close('all')

n = np.random.randn(1000)

m = [m for m in range(len(n))]

plt.figure(figsize=(10, 4))
plt.bar(m, n)
plt.title("Datos de calificaciones")
plt.xlabel("Número de dato")
plt.ylabel("Valor")
plt.show()

plt.figure(figsize=(7, 4))
plt.hist(n, bins=20)
plt.title("Histograma")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.close('all')

x1 = [2, 3, 4]
y1 = [5, 5, 5]

x2 = [1, 2, 3, 4, 5]
y2 = [2, 3, 2, 3, 4]
y3 = [6, 8, 7, 8, 7]

plt.figure(figsize=(7, 4))

plt.scatter(x1, y1, marker="o", label="Grupo 1")
plt.scatter(x2, y2, marker="s", color="blue", label="Grupo 2")
plt.scatter(x2, y3, marker="*", color="green", label="Grupo 3")

plt.title("Scatter Plot Example")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

plt.close('all')

np.random.seed(0)

uniform_data = np.random.rand(10, 50)

plt.figure(figsize=(12, 4))
ax = sns.heatmap(uniform_data)

plt.title("Mapa de calor")
plt.show()
```
